```
price = listingsData['price'].tolist()
plt.hist(price, bins=20)
from numpy import mean,std
mean = mean(price)
std = std(price)
print(mean, std)
#=> 135.86009300746835 19.020565463416645
```

Generate normal distribution, probabilities, and plot it

```
from scipy.stats import norm
distribution = norm(mean, std)

min_price = min(price)
max_price = max(price)
values = list(range(int(min_price), int(max_price)))

probabilities = [distribution.pdf(v) for v in values]

from matplotlib import pyplot
pyplot.hist(price, bins=20, density=True) # , 
pyplot.plot(values, probabilities)
```
